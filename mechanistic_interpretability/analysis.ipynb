```python
df_baseline = pd.read_csv("results/baseline_llama-2-7b.csv", index_col=0)
df_baseline
```

```python
for i in ["One", "Median", "Reset"]:
    df_baseline[f"{i} as %"] = (df_baseline[i] - df_baseline["Clean"]) / df_baseline["Clean"] * 100

df_baseline.drop(columns=["task", "Median", "One", "Reset"]).groupby(["correctness", "order"]).aggregate(["mean"]).round(2)
```
